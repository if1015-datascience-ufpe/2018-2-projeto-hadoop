# Hipótese 2

### Os estados com alto índice de pobreza possuem uma maior mortalidade em incidentes do que estados que possuam baixo índice de pobreza.

A hipótese surgiu pois imaginamos que culturalmente países em crescimento ou desemvolvidos, quando tem um alto índice de pobreza possui uma maior relação no aumento de mortes por incidentes.

#### Carregamento das bases

```python
violencia_df = pd.read_csv('./data-sets/gun-violence-data_01-2013_03-2018.csv')
censo_df = pd.read_csv('./data-sets/acs2015_county_data.csv')
```

### Pré processamento

```python
# Separando os anos em novas colunas
meses = ['January','February','March','April','May','June','July','August','September','October','November','December']
violencia_df['ano'] = violencia_df['date'].apply(lambda x : int(x.split('-')[0]))
violencia_df['mes'] = violencia_df['date'].apply(lambda x : meses[int(x.split('-')[1])-1])
violencia_df['dia'] = violencia_df['date'].apply(lambda x : int(x.split('-')[2]))

# Agrupar por estado
agrupar_df = censo_df.groupby('State').sum()
# Agrupar por população 
agrupar_df = agrupar_df.sort_values('TotalPop', ascending = 0)
violencia_df
agrupar_df
# Além da pobreza, pegamos também as informações com relação 
# a porcentagem das classes raciais que há naquele estado.
agrup = censo_df[['State','County','TotalPop','Poverty','Men','Women','Hispanic','White','Black','Asian','Native']].copy()

agrup.head()
agrup['Hispanic'] = agrup.apply(lambda x : (x['Hispanic'] * x['TotalPop']/100), axis=1)
agrup['White'] = agrup.apply(lambda x : (x['White'] * x['TotalPop']/100), axis=1)
agrup['Black'] = agrup.apply(lambda x : (x['Black'] * x['TotalPop']/100), axis=1)
agrup['Asian'] = agrup.apply(lambda x : (x['Asian'] * x['TotalPop']/100), axis=1)
agrup['Native'] = agrup.apply(lambda x : (x['Native'] * x['TotalPop']/100), axis=1)
agrup.head()

agrup['Poverty_Amount'] = agrup.apply(lambda x: x['TotalPop'] * x['Poverty']/100, axis=1)

pobreza_estado = agrup.groupby('State').sum().reset_index()
pobreza_estado['Poverty'] = pobreza_estado.apply(lambda x: x['Poverty_Amount'] / x['TotalPop'] * 100, axis=1)
pobreza_estado.index = pobreza_estado['State']

#Descartar porto rico por não ser um estado incluso nas outras bases.
pobreza_estado.drop(pobreza_estado[pobreza_estado.index == 'Puerto Rico'].index, axis=0, inplace=True)

#Conversão para porcentagem do total de classes raciais ao agregar informações de todos os estados em uma só.
pobreza_estado['Hispanic'] = pobreza_estado.apply(lambda x : (x['Hispanic'] / x['TotalPop']*100), axis=1)
pobreza_estado['White'] = pobreza_estado.apply(lambda x : (x['White'] / x['TotalPop']*100), axis=1)
pobreza_estado['Black'] = pobreza_estado.apply(lambda x : (x['Black'] / x['TotalPop']*100), axis=1)
pobreza_estado['Asian'] = pobreza_estado.apply(lambda x : (x['Asian'] / x['TotalPop']*100), axis=1)
pobreza_estado['Native'] = pobreza_estado.apply(lambda x : (x['Native'] / x['TotalPop']*100), axis=1)
pobreza_estado.columns = ['State','TotalPop','Poverty','Men','Women','Hispanic (%)','White (%)','Black (%)'\
                         ,'Asian (%)','Native (%)','Poverty_Amount']
etinia = pobreza_estado
```

Será considerado apenas os incidentes acontecidos em 2015, visto que a base censo trata-se do ano de 2015

```python
violencia_2015_df = violencia_df.query('ano==2015', engine='python')
violencia_2015_df = violencia_2015_df.reset_index()
violencia_2015_df.drop(labels='index', inplace=True, axis=1)
#violencia_2015_df = violencia_2015_df.loc[:,['state','n_Killed','gun_type','ano']]
violencia_2015_df.drop(['incident_id', 'date', 'city_or_county', 'address',\
       'n_injured', 'incident_url', 'source_url',\
       'incident_url_fields_missing', 'incident_characteristics','congressional_district', 'gun_stolen',\
       'location_description', 'n_guns_involved', 'notes',\
       'participant_age', 'participant_age_group', 'participant_gender',\
       'participant_name', 'participant_relationship', 'participant_status',\
       'participant_type', 'sources', 'state_house_district',\
       'state_senate_district', 'mes', 'dia'], axis=1, inplace=True)
man = violencia_2015_df.loc[:,['state','n_killed']].groupby('state')
violencia_status_df = pd.DataFrame()
violencia_status_df['QTD_incidentes'] = man.size()
violencia_status_df['n_mortos'] = man.sum()
violencia_status_df
```

### Conclusão

Através do mapa acima podemos visualizar uma correlação da renda com os trabalhos formais ou não, dessa forma esta constatação nos permite inferir, com o auxílio dos gráficos abaixo, que com a correlação existente entre desemprego, números de mortos nos incidentes e a taxa de pobreza em cada estado, a hipotese seja válida. Ou seja, A probabilidade de um cidadão estar entre os trabalhadores pobres varia muito de acordo com a ocupação. Os trabalhadores em profissões que exigem educação superior e se caracterizavam por rendimentos relativamente altos - como ocupações administrativas, profissionais e afins - eram menos propensos a serem classificados como trabalhadores pobres. Por exemplo, 1,8% dos ocupantes de cargos gerenciais, profissionais e relacionados estavam entre os trabalhadores pobres em 2015. Em contraste, os indivíduos empregados em ocupações que normalmente não exigem altos níveis de educação e que são caracterizados por ganhos relativamente baixos eram mais prováveis. estar entre os trabalhadores pobres. Por exemplo, 11,6% dos trabalhadores de serviços que estavam na força de trabalho por pelo menos 27 semanas foram classificados como trabalhadores pobres em 2015. De fato, ocupações de serviço, com 3,0 milhões de trabalhadores pobres, representavam 38% de todos os classificados como trabalhadores pobres. Entre aqueles empregados em recursos naturais, construção e ocupações de manutenção, 6,9% foram classificados como trabalhadores pobres. Dentro deste grupo de ocupação, 14,1% dos trabalhadores nas ocupações de agricultura, pesca e silvicultura estavam entre os trabalhadores pobres.Segundo as informações do [Bureau_of_labor_statistic](https://www.bls.gov/opub/reports/working-poor/2015/home.htm#table1).

Tudo isso fica evidente quanto analizamos o estado da California que em 2015 foi o único estado observado com alta porcentagem de pobreza e desemprego proximo ao 10%, devido ao fato de, mesmo estando nesse ano com um PIB considerável e com uma poplução multiétnica considerável,  segundo o [BBC](https://www.bbc.com/portuguese/internacional-36731177) 
houve alguns Problemas visto que, alguns acreditam que os números que levaram o Estado a estar entre as seis grandes economias do mundo foram mais um acidente contábil proveniente da valorização do dólar diante de outras moedas do que qualquer coisa.

Outros indicam que essa mesma valorização faria as indústrias da região menos competitivas diante de concorrentes como o México, fazendo com que os empregos acabem indo para outro país.

E também há quem se queixe da regulamentação estatal, dizendo que ela é hostil aos negócios.

Em 2015, a gigante petroleira Oxy, que estava há quase um século em Los Angeles, anunciou que mudaria sua sede para Houston, no Texas.

```python
pobreza_df = censo_df[['State','Unemployment' ,'County', 'TotalPop', 'Poverty']].copy()
pobreza_df['QTD_pobreza'] = pobreza_df.apply(lambda r: r['TotalPop'] * r['Poverty']/100, axis=1)
pobreza_df['QTD_desempregados'] = pobreza_df.apply(lambda r: r['TotalPop'] * r['Unemployment']/100, axis=1)

pobreza_estado_df = pobreza_df.groupby('State').sum().reset_index()
pobreza_estado_df['P_desempregados'] = pobreza_estado_df.apply(lambda r: r['QTD_desempregados'] / r['TotalPop'] * 100, axis=1)
pobreza_estado_df['P_Pobreza'] = pobreza_estado_df.apply(lambda r: r['QTD_pobreza'] / r['TotalPop'] * 100, axis=1)
pobreza_estado_df.index = pobreza_estado_df['State']
#será desconsiderado porto rico por não fazer parte do EUA, e não ser usado em outras bases.
pobreza_estado_df.drop(pobreza_estado_df[pobreza_estado_df.index == 'Puerto Rico'].index, axis=0, inplace=True)
pobreza_estado_df['P_desempregados'].max()
pobreza_estado_df
```
